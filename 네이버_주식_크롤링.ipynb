```python
# 1. 라이브러리 호출
import requests
from bs4 import BeautifulSoup as bs
```

```python
# 2. 데이터 요청 단계
# requests.get("URL 주소")
req = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1")
html = req.text # html정보만을 꺼내오기 위해 .test
soup = bs(html,"lxml") # 컴퓨터가 이해할 수 있도록 html -> lxml
```

```python
print(soup)
```

```python
# 3. 데이터 추출(파싱)단계
# F12 -> 주식 종목 선택 -> 마우스 우클릭 -> copy -> copy selector
stockContents = soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr") 
print(stockContents)
```

```python
# 코스피 1페이지만 보기
for stockContent in stockContents:
  try: # 일단 한 번 해봐
    stockRank = stockContent.select_one("td.no").text
    stockName = stockContent.select_one("td:nth-of-type(2) > a").text
    stockPrice = stockContent.select_one("td:nth-of-type(3)").text
    stockCap = stockContent.select_one("td:nth-of-type(7)").text
    # print(stockRank)
    # print(stockName)
    # print(stockPrice)
    # print(stockCap)
    print(f"{stockRank}등 {stockName} 종목의 현재가는 {stockPrice}원 이고, 시가총액은 {stockCap}억 원 입니다.")
  except AttributeError: # AttributeError가 나오면 그냥 넘어가줘
    continue
```
